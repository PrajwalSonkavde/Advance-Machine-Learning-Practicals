# splitting the data

```python
x=df.drop(['target'],axis=1)
y=df['target']
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

# Normalizing data

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
cols = x_train.columns
sc_x_train = sc.fit_transform(x_train)
sc_x_train = pd.DataFrame(sc_x_train , columns = cols)
sc_x_test = sc.fit_transform(x_test)
sc_x_test = pd.DataFrame(sc_x_test , columns = cols)
```

# Comparing accuracy of different model

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
rf = RandomForestClassifier()
rf.fit(sc_x_train , y_train)
pred = rf.predict(sc_x_test)
print(accuracy_score(pred , y_test))
```

### Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gboosting = GradientBoostingClassifier()
gboosting.fit(sc_x_train , y_train)
pred = gboosting.predict(sc_x_test)
print(accuracy_score(pred , y_test))
```

### XGBoost Classifier

```python
from xgboost import XGBClassifier

model7 = XGBClassifier()
model7.fit(x_train, y_train)
y_pred7 = model7.predict(x_test)
print(accuracy_score(y_test, y_pred7))
```

### KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(sc_x_train , y_train)
pred = knn.predict(sc_x_test)
print(accuracy_score(y_test, pred))
```

### Naive bayes Classifier

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(sc_x_train , y_train)
pred = nb.predict(sc_x_test)
print(accuracy_score(y_test, pred))
```

### Majority Vote Classifier

```python
models = []
rf = RandomForestClassifier()
models.append(rf)
rf = RandomForestClassifier()
models.append(rf)
rf = RandomForestClassifier()
models.append(rf)
model7 = XGBClassifier()
models.append(model7)
model7 = XGBClassifier()
models.append(model7)
knn = KNeighborsClassifier()
models.append(knn)
knn = KNeighborsClassifier()
models.append(knn)
knn = KNeighborsClassifier()
models.append(knn)
knn = KNeighborsClassifier()
models.append(knn)
nb = GaussianNB()
models.append(nb)
nb = GaussianNB()
models.append(nb)
nb = GaussianNB()
models.append(nb)
nb = GaussianNB()
models.append(nb)


models
```

```python
import numpy as np
import pandas as pd

class MajorityVoteClassifier:
    def __init__(self, classifiers):
        self.classifiers = classifiers
        
    def fit(self, X, y):
        for classifier in self.classifiers:
            classifier.fit(X, y)
            
    def predict(self, X):
        predictions = pd.DataFrame()
        for i,classifier in enumerate(self.classifiers):
            pred = classifier.predict(X)
            predictions["col{}".format(i)] = pred
        return predictions.mode(axis=1)
```

```python
mvc = MajorityVoteClassifier(models)
mvc.fit(sc_x_train , y_train)
pred = mvc.predict(sc_x_test)
print(accuracy_score(y_test, pred))
```
